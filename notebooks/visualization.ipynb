```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas
import os
```

## Visualization for IMDB-Binary

```python
imdb_binary = os.path.join('../data/', 'IMDB-BINARY')

imdb_binary_A = os.path.join(imdb_binary, 'IMDB-BINARY_A.txt')
imdb_binary_A_data = pandas.read_csv(imdb_binary_A, sep='\t', header=None)

imdb_binary_graph1_edge = imdb_binary_A_data[:292]
imdb_binary_graph1_edge.columns = ['edge']
imdb_binary_graph1_edge['edge'] = imdb_binary_graph1_edge['edge'].str.split(', ')
imdb_binary_graph1_edge
```

```python
# create a graph
imdb_binary_graph1 = nx.Graph()

# # add nodes
# for i in range(20):
#     imdb_binary_graph1.add_node(i)

# add edges
for i in range(292):
    imdb_binary_graph1.add_edge(imdb_binary_graph1_edge['edge'][i][0], imdb_binary_graph1_edge['edge'][i][1])
```

```python
# draw the graph
nx.draw(imdb_binary_graph1, with_labels=True)  
plt.title('IMDB-BINARY Graph 1')
plt.show()
```

```python
imdb_binary_graph2_edge = imdb_binary_A_data[292:804]
imdb_binary_graph2_edge.columns = ['edge']
imdb_binary_graph2_edge['edge'] = imdb_binary_graph2_edge['edge'].str.split(', ')
imdb_binary_graph2_edge
```

```python
imdb_binary_graph2 = nx.Graph()

for i in range(292, 804):
    imdb_binary_graph2.add_edge(imdb_binary_graph2_edge['edge'][i][0], imdb_binary_graph2_edge['edge'][i][1])

nx.draw(imdb_binary_graph2, with_labels=True)
plt.title('IMDB-BINARY Graph 2')
plt.show()
```
